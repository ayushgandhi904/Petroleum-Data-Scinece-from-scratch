##### Random number generation

Data Distribution

```python
# Random distribution =======>>>> A random distribution is a set of random numbers that follows that follow a certain probability
# density function
```

```python
# Probability Density function----- A function that describes the continuous probability. i.e. probability of all values in an array
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# choice ==> we can give probablity for each value
```

```python
x = np.random.choice([1, 2, 3, 4, 5], p = [0.1, 0.2, 0.3, 0.25, 0.15], size = 100)
```

```python
x
```

```python
plt.hist(x, bins = 5)
```

```python
sns.displot(x, bins = 5, kde = True) #kde = kernel density plot
```

```python
sns.distplot(x)
```

## Normal Distribution

```python
x = np.random.normal(loc = 10, scale = 5, size = (1000))
y = np.random.normal(loc = 5, scale = 10, size = (1000))
```

```python
sns.distplot(x)
sns.distplot(y)
```
